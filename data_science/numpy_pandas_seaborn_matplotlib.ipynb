# Matplotlib.pyplot

```python
# figure is the canvas
# ax is a plot on the canvas
(fig, ax) = plt.subplots()

# for different colours per 'cut' category:
# group the datafram by 'cut' and loop over it
for category, group_data in df.groupby("cut"):
    ax.scatter(group_data["carat"], group_data['price'], label=category)
    
plt.title("The price of diamonds")
plt.xlabel("Weight (carats)")
plt.ylabel("Price ($)")
plt.show()
```

```python
# multiple axes can be put on a canvas (2 row, 2 col, 5x5 each)
# (colours not added)
(fig, axes) = plt.subplots(2, 2, figsize=(10,10))
(ax1, ax2, ax3, ax4) = axes.flatten()

# can use seaborn or matplotlib
# can use set_xlabel, sety_label or just set
ax1.scatter(df["carat"], df['price'])
ax1.set_xlabel("Weight (carats)")
ax1.set_ylabel("Price ($)")
ax1.set_title('matplotlib')

# seaborn + set is the most efficient way
sns.scatterplot(data=df, x='carat', y='price', ax=ax2)
ax2.set(title='seaborn', xlabel='Weight (carats)', ylabel='Price ($)')

ax3.hist(df["carat"])
ax3.set(title='matplotlib', xlabel='Weight (carats)', ylabel='Price ($)')

sns.histplot(data=df, x='carat', ax=ax4)
ax4.set(title='seaborn', xlabel='Weight (carats)', ylabel='Price ($)')

plt.suptitle("The price of diamonds")
plt.tight_layout() # prevents x,y labels from overlapping
plt.show()
```

## Mathematical 2D/3D plotting

```python
n = 500
x = np.random.randn(n)
y = np.random.randn(n)
sns.scatterplot(x=x, y=y)
plt.title('Mathematical plotting')
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
# define a function to seperate the values
def f(x):
    return x**2 - 1

# values above the function are green (#0f0) -> y more than f(x)
# values below the function are red (#f00) -> y less than f(x)
colours = np.where(y > f(x), '#0f0', '#f00')
sns.scatterplot(x=x, y=y, c=colours)
plt.title('Function Classification')
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
# create a vectors field of the points in direction (u,v)
x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)
(x, y) = np.meshgrid(x, y)

u = 1
v = -1
plt.quiver(x, y, u, v)
plt.title('Vector Fields')
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
u = x
v = y
plt.quiver(x, y, u, v)
plt.title('Vector Field')
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
# f(x,y) = (y / sqrt(x^2 + y^2) - (x / sqrt(x^2 + y^2)
u = y / np.sqrt(x**2 + y**2)
v = -x / np.sqrt(x**2 + y**2)
plt.quiver(x, y, u, v)
plt.title('Vector Field')
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```
